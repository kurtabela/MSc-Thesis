```python
import math
import langid
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch
import numpy as np
import tensorflow as tf
import tensorflow_text
import math
from joblib import Parallel, delayed


tokenizer_mt = AutoTokenizer.from_pretrained("MLRS/BERTu", use_fast=True)
model_mt = AutoModelForMaskedLM.from_pretrained("MLRS/BERTu")
# set device to GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# put model on GPU (is done in-place)
model_mt.to(device)
model_mt.eval()

```

```python
tokenizer_en = AutoTokenizer.from_pretrained("bert-base-uncased", use_fast=True)
model_en = AutoModelForMaskedLM.from_pretrained("bert-base-uncased")
# set device to GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
# put model on GPU (is done in-place)
model_en.to(device)
model_en.eval()

softmax = torch.nn.Softmax(dim=0)

#en_mt = tf.saved_model.load('translator_for_confidence')


```

```python


print(device)
def lm_en_prob(sentence):

    tokenize_input = tokenizer_en(sentence, padding=True, return_tensors='pt')["input_ids"].cuda()

    with torch.no_grad():
        tokenize_input.to(device)
        loss = model_en(tokenize_input.cuda())[0]
    total_scores = []
    length_of_input = len(tokenize_input[0])
    for j, individual_score in enumerate(loss):
        print(j)
        
        scores = []
        
        length_of_input = len(tokenize_input[j])
        softmax_of_sent = softmax(loss[j])
        for i in range(length_of_input):
            scores.append(torch.log(softmax_of_sent[i])[tokenize_input[j, i]])
        total_scores.append(scores)
    return total_scores


def lm_mt_prob(sentence):
    tokenize_input = tokenizer_en(sentence, padding=True, return_tensors='pt')["input_ids"].cuda()

    with torch.no_grad():
        tokenize_input.to(device)
        print("feeding to model...")
        loss = model_mt(tokenize_input)[0]
        print("done.")
    total_scores = []
    for j, individual_score in enumerate(loss):
        print(j)
        
        scores = []
        
        length_of_input = len(tokenize_input[j])
        softmax_of_sent = softmax(loss[j])
        for i in range(length_of_input):
            scores.append(torch.log(softmax_of_sent[i])[tokenize_input[j, i]])
        total_scores.append(scores)
    return total_scores


def mt_en_prob(sentence_en, sentence_mt):
    return [0]
    # return en_mt(tf.constant(sentence_en), tf.constant(sentence_mt))



def mt_mt_prob(sentence_mt, sentence_en):
    return [0]
    # return en_mt(tf.constant(sentence_en), tf.constant(sentence_mt))



def sentence_score(sentence_en, sentence_mt, score):
    lm_en = lm_en_prob(sentence_en)
    print("FINISHED EN")
    lm_mt = lm_mt_prob(sentence_mt)
    print("FINISHED MT")

    for i, sentence in enumerate(sentence_en):
        if score[i] == 0:
            mt_en = mt_en_prob(sentence, sentence_mt[i])
            mt_mt = mt_mt_prob(sentence_mt[i], sentence)

            lm_scores = ((1 / len(lm_en[i])) * sum(lm_en[i])) + (
                    (1 / len(lm_mt[i])) * sum(lm_mt[i]))
            mt_scores = ((1 / len(mt_en)) * sum(mt_en)) + (
                    (1 / len(mt_mt)) * sum(mt_mt))

            # print(lm_scores)
            score[i] = (1 / 4) * (lm_scores + mt_scores)
    return score


def inner_for(i,sentences_en, sentences_mt,lines_to_remove, length_sentences_en):
    torch.cuda.empty_cache() # PyTorch thing
    print(i*batch_size)
    if (i*batch_size)+batch_size >= length_sentences_en:
        sentence_en = sentences_en[(i*batch_size):length_sentences_en]
        sentence_mt = sentences_mt[(i*batch_size):length_sentences_en]
        scores = np.zeros(length_sentences_en-i*batch_size)
    else:
        sentence_en = sentences_en[i*batch_size:(i*batch_size)+batch_size]
        sentence_mt = sentences_mt[i*batch_size:(i*batch_size)+batch_size]
        scores = np.zeros(batch_size)

    for j, sentence in enumerate(sentence_en):
        response = langid.classify(sentence)
        if response[0] != "en":
            # print("Not English: " + sentence_en)
            scores[j] = -1

    for j, sentence in enumerate(sentence_mt):
        response = langid.classify(sentence)
        if response[0] != "mt":
            # print("Not Maltese: " + sentence_mt)
            scores[j] = -1

    # for index in sorted(wrong_language, reverse=True):
    #     del sentence_mt[index]
    #     del sentence_en[index]


    scores = sentence_score(sentence_en, sentence_mt, scores)
    print(scores)
    for score in scores:
        if score < -7:
            lines_to_remove.append(i)
    return scores

```
